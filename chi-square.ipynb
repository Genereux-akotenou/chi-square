#### SIMULATION

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
# Paramètres de la distribution chi-carré
df = 5  # Degrés de liberté

# Génération d'un échantillon de variables aléatoires chi-carré
sample_size = 1000
chi_square_sample = np.random.chisquare(df, size=sample_size)

# Affichage de l'histogramme de l'échantillon
plt.hist(chi_square_sample, bins=20, density=True, alpha=0.6, color='b', label='Chi-Square Sample')
plt.title(f'Échantillon de la distribution chi-carré (df={df})')
plt.xlabel('Valeur')
plt.ylabel('Fréquence')
plt.legend()
plt.grid(True)

plt.show()
```

#### DENSITÉ DE PROBABILITÉ PDF

```python
# Calcul de la PDF de la distribution chi-carré
x1 = np.linspace(0, 20, 1000)
pdf1 = chi2.pdf(x1, 5)

x2 = np.linspace(0, 20, 1000)
pdf2 = chi2.pdf(x2, 6)

x3 = np.linspace(0, 20, 1000)
pdf3 = chi2.pdf(x3, 8)

# Affichage de l'histogramme de l'échantillon
plt.hist(chi_square_sample, bins=20, density=True, alpha=0.6, color='b', label='Chi-Square Sample')

plt.plot(x1, pdf1, 'r', label=f'Chi-Square PDF (df={df})')
plt.plot(x2, pdf2, 'y', label=f'Chi-Square PDF (df={df})')
plt.plot(x3, pdf3, 'y', label=f'Chi-Square PDF (df={df})')


plt.title(f'Échantillon et densité de probabilité de la distribution chi-carré (df={df})')
plt.xlabel('Valeur')
plt.ylabel('Fréquence / Densité de probabilité')
plt.legend()
plt.grid(True)

plt.show()
```

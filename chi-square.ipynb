```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramètres de la distribution chi-carré
df = 3  # Degrés de liberté

# Génération d'un échantillon de variables aléatoires chi-carré
sample_size = 1000
chi_square_sample = np.random.chisquare(df, size=sample_size)

# Affichage de l'histogramme de l'échantillon
plt.hist(chi_square_sample, bins=20, density=True, alpha=0.6, color='b', label='Chi-Square Sample')
plt.title(f'Échantillon de la distribution chi-carré (df={df})')
plt.xlabel('Valeur')
plt.ylabel('Fréquence')
plt.legend()
plt.grid(True)

plt.show()
```
